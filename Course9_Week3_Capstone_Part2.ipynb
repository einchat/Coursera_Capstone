```python
import numpy as np
import pandas as pd
```

```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
webpage = requests.get(url)
```

```python
webpage
```

```python
# get the webpage data
data = webpage.text
```

```python
soup= BeautifulSoup(data)
```

```python
# find table tag in the webpage data
# checked that the first table contains the necessary data
table = soup.find_all('table')[0]
```

```python
df = pd.read_html(str(table)) #Read HTML tables into a list of DataFrame objects
```

```python
df[0].head()
```

```python
Nbhd_df = df[0][1:]
```

```python
Nbhd_df.head()
```

```python
Nbhd_df.columns = ['Postcode', 'Borough','Neighbourhood']
```

```python
Nbhd_df.head(2)
```

```python
# checking for 'Not assigned' boroughs
Nbhd_df[Nbhd_df['Borough']=='Not assigned']
```

```python
Nbhd_df.shape
```

```python
# droping these line items as per instruction
Nbhd_df.drop( Nbhd_df[Nbhd_df['Borough']=='Not assigned'].index, axis=0, inplace=True)
```

```python
Nbhd_df.head()
```

```python
# verifying
Nbhd_df[Nbhd_df['Neighbourhood']=='Not assigned']
```

```python
Nbhd_df.groupby('Postcode').count()
```

```python
# lamda function to concatenate the neighborhoods with comma as separator
Nbhd_group= pd.DataFrame(Nbhd_df.groupby('Postcode')['Neighbourhood'].apply(lambda x:  ', '.join(x)))
```
